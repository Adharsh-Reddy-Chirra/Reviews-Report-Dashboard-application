```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Load the mini dataset
mini_dataset_path = "C:\\Users\\adhar\\Downloads\\Final_Project_Phase_3(1)\\Mini_Dataset_Amazon_Appliances_Reviews.csv"
mini_df = pd.read_csv(mini_dataset_path)

# Drop rows with missing values in the 'reviewText' column
mini_df.dropna(subset=['reviewText'], inplace=True)

# Convert years to string format
mini_df['year'] = pd.to_datetime(mini_df['reviewTime']).dt.year.astype(str)

# Create a dictionary to store similar reviews for each year
similar_reviews_by_year = {}

# Iterate over the years and find similar reviews
for year in ['2011', '2012', '2013', '2014']:
    # Filter reviews for the current year
    reviews_year = mini_df[mini_df['year'] == year]['reviewText'].tolist()
    
    # Create TF-IDF vectors for reviews
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(reviews_year)
    
    # Calculate cosine similarity between reviews
    cosine_similarities = cosine_similarity(tfidf_matrix, tfidf_matrix)
    
    # Get indices of top 3 most similar reviews for each review
    similar_reviews_indices = [sorted(range(len(row)), key=lambda x: row[x], reverse=True)[1:4] for row in cosine_similarities]
    
    # Store top 3 most similar reviews
    similar_reviews = [[reviews_year[idx][:100] for idx in indices[:3]] for indices in similar_reviews_indices]
    
    # Add to dictionary
    similar_reviews_by_year[year] = similar_reviews

# Print the top 3 most similar reviews for each year
for year, reviews in similar_reviews_by_year.items():
    print(f"Top 3 most similar reviews for {year}:")
    for idx, review_set in enumerate(reviews, start=1):
        print(f"Similar Review {idx}:")
        for review in review_set:
            print(review)
        print()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Load the mini dataset
mini_dataset_path = "C:\\Users\\adhar\\Downloads\\Final_Project_Phase_3(1)\\Mini_Dataset_Amazon_Appliances_Reviews.csv"
mini_df = pd.read_csv(mini_dataset_path)

# Drop rows with missing values in the 'reviewText' column
mini_df.dropna(subset=['reviewText'], inplace=True)

# Create TF-IDF vectors for reviews
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(mini_df['reviewText'])

# Calculate cosine similarity between reviews
cosine_similarities = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Get indices of top 3 most similar reviews for each review
similar_reviews_indices = [sorted(range(len(row)), key=lambda x: row[x], reverse=True)[1:4] for row in cosine_similarities]

# Store top 3 most similar reviews
similar_reviews = [[mini_df.iloc[idx]['reviewText'] for idx in indices] for indices in similar_reviews_indices]

# Print the top 3 most similar reviews
for idx, review_set in enumerate(similar_reviews, start=1):
    print(f"Similar Review {idx}:")
    for review in review_set:
        print(review[:100])  # Print only the first 100 characters of the review
    print()
```

```python
import pandas as pd

# Load the mini dataset containing reviews
mini_dataset_path = "C:\\Users\\adhar\\Downloads\\Final_Project_Phase_3(1)\\Mini_Dataset_Amazon_Appliances_Reviews.csv"
mini_dataset = pd.read_csv(mini_dataset_path)

# Step 1: Drop rows with missing values in the 'reviewText' column
mini_dataset.dropna(subset=['reviewText'], inplace=True)

# Step 2: Filter reviews mentioning "value"
value_reviews = mini_dataset[mini_dataset['reviewText'].str.contains('value', case=False)]

# Step 3: Calculate the average rating for each reviewer
reviewers_avg_rating = value_reviews.groupby('reviewerName')['overall'].mean()

# Step 4: Filter reviewers with an average rating higher than 4
selected_reviewers = reviewers_avg_rating[reviewers_avg_rating > 4]

# Display the selected reviewers
print("Reviewers with reviews mentioning 'value' and average ratings higher than 4:")
print(selected_reviewers)
```
